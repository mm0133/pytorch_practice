```python
import torch

X=torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
Y=torch.FloatTensor([[0],[1],[1],[0]])

linear= torch.nn.Linear(2,1,bias=True)
sigmoid = torch.nn.Sigmoid()

model =torch.nn.Sigmoid()
model= torch.nn.Sequential(linear, sigmoid)
criterion=torch.nn.BCELoss()
optimizer=torch.optim.SGD(model.parameters(), lr=1)

for step in range(10001):
  optimizer.zero_grad()
  hypothesis=model(X)
  cost=criterion(hypothesis,Y)
  cost.backward()
  optimizer.step()
  if step% 100 ==0:
    print(step, cost.item())

```

```python
#프레임워크 함수를 적게쓰고 완성해보기
import torch
torch.manual_seed(2)

X=torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
Y=torch.FloatTensor([[0],[1],[1],[0]])

W1= torch.randn(2,2)
b1=torch.randn(2)
W2=torch.randn(2,1)
b2=torch.randn(1)
print(W1,b1)
lr=1
def sigmoid(x):
  return 1.0/(1.0+torch.exp(-x))

def sigmoid_prime(x):
  return sigmoid(x)*(1-sigmoid(x))

for step in range(10001):
  l1=torch.add(torch.matmul(X,W1),b1)
  a1=sigmoid(l1)
  l2=torch.add(torch.matmul(a1,W2),b2)
  Y_pred=sigmoid(l2)
  cost = -torch.mean(Y*torch.log(Y_pred)+(1-Y)*torch.log(1-Y_pred))

  
  d_Y_pred = (Y_pred-Y)/(Y_pred*(1.0 - Y_pred)+1e-7)
  d_l2 =d_Y_pred*sigmoid_prime(l2)
  d_b2=d_l2
  d_W2=torch.matmul(torch.transpose(a1,0,1),d_b2)

  d_a1 =torch.matmul(d_b2, torch.transpose(W2,0,1))
  d_l1= d_a1*sigmoid_prime(l1)
  d_b1=d_l1
  d_W1=torch.matmul(torch.transpose(X,0,1),d_b1)
  W1-=lr*d_W1
  b1-=lr*torch.mean(d_b1,0)
  W2-=lr*d_W2
  b2-=lr*torch.mean(d_b2,0)
  if step%100==0:
    print(step,cost.item())

```

```python
#프레임워크사용
import torch
torch.manual_seed(2)

X=torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
Y=torch.FloatTensor([[0],[1],[1],[0]])

l1=torch.nn.Linear(2,2,bias=True)
l2=torch.nn.Linear(2,1,bias=True)
sigmoid=torch.nn.Sigmoid()

model=torch.nn.Sequential(l1,sigmoid,l2,sigmoid)
criterion=torch.nn.BCELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=1)

for step in range(10001):
  optimizer.zero_grad()
  hypothesis=model(X)
  cost=criterion(hypothesis,Y)
  cost.backward()
  optimizer.step()
  if step%100==0:
    print(step,cost.item())
```

```python
# 레이어 늘리기

import torch
torch.manual_seed(2)

X=torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
Y=torch.FloatTensor([[0],[1],[1],[0]])

N=1
layers=[torch.nn.Linear(2,10,bias=True),torch.nn.Sigmoid()]
for i in range(N):
  layers.append(torch.nn.Linear(10,10,bias=True))
  layers.append(torch.nn.Sigmoid())
layers.append(torch.nn.Linear(10,1,bias=True))
layers.append(torch.nn.Sigmoid())

model=torch.nn.Sequential(*layers)
criterion=torch.nn.BCELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=1)

for step in range(10001):
  optimizer.zero_grad()
  hypothesis=model(X)
  cost=criterion(hypothesis,Y)
  cost.backward()
  optimizer.step()
  if step%100==0:
    print(step,cost.item())
```
